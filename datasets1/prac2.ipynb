```python
years= range(2000,2012)
apples = [0.895,0.91,0.919,0.926,0.929,0.931, 0.934,0.936,0.937,0.938,0.939,0.940]
orange = [0.962,0.941,0.930,0.923,0.918,0.908,0.907,0.904,0.901,0.902,0.903,0.904]
```

```python
# line graph
fig, axes= plt.subplots(2,3, figsize=(12,9))
axes[0,0].plot(years, orange, 'o--r')
axes[0,0].plot(years,apples, 's--b')
axes[0,0].set_title('crop yields')
axes[0,0].set_xlabel('year')
axes[0,0].set_ylabel('yield')
axes[0,0].legend(['apples', 'orange'])
plt.tight_layout(pad=2)

axes[0,1].set_title('sepal width vs sepal length')
sns.scatterplot(x=flowers.sepal_length,
                y= flowers.sepal_width, 
                hue=flowers.species, 
                s=100, 
                ax =axes[0,1])

axes[0,2].set_title('flight traffick')
sns.heatmap(flight, cmap= 'Blues', ax=axes[0,2])

axes[1,0].set_title('restaurant bills')
sns.barplot(x='day',
            y='total_bill', 
            hue = 'sex',
            data=tip, 
            ax=axes[1,0])

axes[1,1].set_title('distribution of sepal width')
sns.histplot(x=setosa.sepal_width,
             y= versicolor.sepal_width,
             hue=virginica.species, bins=np.arange(2,5,0.25),
                       stacked=True, ax=axes[1,1])
```

# data preparation and cleaning
next step is preprocessing and cleaning the datasets

while the survey contains a wealth of information, we will limit our analysis to
1. demographics of the survey respondents and the global programming communities
2. distribution of programming skills, experience and preference
3. employment related information, preference and opinions

```python
selected_columns=[
# select a subset from the column datasets
# demographics
'Country',
'Age',
'Gender',
'EdLevel',
'UndergradMajor',
# programming experience
'Hobbyist',
'Age1stCode',
'YearsCode',
'YearsCodePro',
'LanguageWorkedWith',
'LanguageDesireNextYear',
'NEWLearn',
'NEWStuck',
# employment
'Employment',
'DevType',
'WorkWeekHrs',
'JobSat',
'JobFactors',
'NEWOvertime',
'NEWEdImpt',
]
```

```python
# convert the string object to numeric columns 
# by ignoring the sting and changing it to nan values
surv['Age1stCode']= pd.to_numeric(surv.Age1stCode, errors = 'coerce')
surv['YearsCode']= pd.to_numeric(surv.YearsCode, errors = 'coerce')
surv['YearsCodePro']= pd.to_numeric(surv.YearsCodePro, errors = 'coerce')
```

```python
surv.describe()
```

there seems to be a problemin the age column, as the minimum valis is 1 and max value is 279.
this is a common error in survey:responses may contain invalid 
dues to accidental or intentional errors while responding. 
a simple fix would be to ignore the rows where the value in the age column is
higher than 100 years or lower than 10 years as invalid survey responses.
this can be done using the drop function  

```python
surv.drop(surv[surv.Age<10].index, inplace =True)
surv.drop(surv[surv.Age>100].index, inplace= True)
```

we have the same issues with the workweekhrous

```python
surv.drop(surv[surv.WorkWeekHrs<140].index, inplace = True)
```
